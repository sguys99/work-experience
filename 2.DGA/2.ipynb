3) eq_num(설비번호) 컬럼 추가  
- **향 후:** 결측치 1292건 최종확인(실제 없는 것인지 누락된 것인지 확인)

```python
def return_num(name):
    if '#' in name:
        tmp = name.split('#')[1].split(' ')[0]
        tmp = '#'+tmp
        
        if (len(tmp)==2)&(tmp[1].isdigit()):
            return tmp
        elif tmp[2].isdigit():
            return tmp[0:2] + '-'+tmp[2]
        elif tmp[2]=='-':
            return tmp[0:4]
        else:
            return tmp[0:2]
    else:
        return 'NA'
```

```python
df['eq_num'] = df['name'].apply(return_num)
```

4) eq_phase(상) 컬럼 추가  
- **향 후:** 결측치 최종확인(실제 없는 것인지 누락된 것인지 확인)

```python
def return_phase(name):
        
    if ' A' in name:
        if 'A-1' in name:
            return 'A-1'
        elif 'A-2' in name:
            return 'A-2'
        elif 'A-3' in name:
            return 'A-3'
        else:
            return 'A'
   
    elif ' B' in name:
        if 'B-1' in name:
            return 'B-1'
        elif 'B-2' in name:
            return 'B-2'
        elif 'B-3' in name:
            return 'B-3'
        else:
            return 'B'
        
    elif ' C' in name:
        if 'C-1' in name:
            return 'C-1'
        elif 'C-2' in name:
            return 'C-2'
        elif 'C-3' in name:
            return 'C-3'
        else:
            return 'C'
    
    elif 'A-' in name:
        if 'A-1' in name:
            return 'A-1'
        elif 'A-2' in name:
            return 'A-2'
        elif 'A-3' in name:
            return 'A-3'
        else:
            return 'A'
        
    elif 'B-' in name:
        if 'B-1' in name:
            return 'B-1'
        elif 'B-2' in name:
            return 'B-2'
        elif 'B-3' in name:
            return 'B-3'
        else:
            return 'B'    
        

    elif 'C-' in name:
        if 'C-1' in name:
            return 'C-1'
        elif 'C-2' in name:
            return 'C-2'
        elif 'C-3' in name:
            return 'C-3'
        else:
            return 'C' 
        
    elif '상' in name:
        if 'A상' in name:
            return 'A'
        elif 'B상' in name:
            return 'B'
        elif 'C상' in name:
            return 'C'
        else:
            return 'NA'
    else:
        return 'NA'
```

2) volt_by_name : name 컬럼에서 전압정보 추출하기  
- **결측치 :** 142개 (결측치는 모두 정주 변전소의 것임. 확인하여 입력필요)('20.03.18)
- **검토사항1 :** 66kV가 일반적인 값인지 확인 필요('20.03.18)
- **검토사항2 :** name 컬럼의 전압 값이 맞는지, volt 컬럼이 맞는지 확인필요('20.03.18)
    * volt vs. volt_by_name 차이가 나는 데이터를 저장해서 살펴본 결과, 명확히 이름상의 명시된 kV를 틀린 경우가 많음 volt_by_name가 더 신뢰성이 높음('20.03.31)
    * 따라서, 상단파트에서 크로스체크하면서 채운 None 값 부위도 신뢰성이 떨어질 수 있음('volt_vs_name_volt.xlsx' 파일로 전력연에 확인 요청 필요)('20.03.31)
- **향 후:** volt 컬럼과 volt_by_name 컬럼 불일치 처리

```python
def return_volt_by_name(row):
    data = str(row['name']).upper()
    
    if '765' in data:
        return '765'
    elif '345' in data:
        return '345'
    elif '154' in data:
        return '154'
    elif '66' in data:
        return '66'
    elif ('23KV'or '23' or '22.9kV' or '22.9') in data:
        if '/'in data:
            tmp = data.split('/')[0]
            if tmp.isdigit():
                return tmp
            
            else:
                return '22.9'
            
        else:
            return '22.9'
    else:
        return row['volt']
```

```python
df['volt_by_name'] = df[['name', 'volt']].apply(return_volt_by_name, axis = 1)
```
