```python
def load_data(inputPath, dtype='pickle'):
    
    if dtype == 'pickle':
        df = pd.read_pickle(inputPath)
    elif dtype == 'csv':
        df = pd.read_csv(inputPath)
    else:
        return np.nan
    
    assert len(df.columns)==21, 'Check column length'
    
    cols = ['1차사업소명','2차사업소명','설비명','제작사','전압','제작년도',
            '분석일자','설비구분', '분석구분', 
            'H2', 'C2H2', 'C2H4', 'CH4', 'C2H6', 'C3H8', 'TCG',
            '판정(가스) 명', 'CO', 'CO2', '판정(COCO2) 명','특기사항']
    
    assert cols == [i.strip() for i in list(df.columns)], 'Check column names'
    
    
    df.columns = ['office1', 'office2', 'name', 'maker', 'volt', 'mfg_year',
                  'sample_date', 'operation', 'anal_type',
                  'H2', 'C2H2', 'C2H4', 'CH4', 'C2H6', 'C3H8', 'TCG',
                  'r1(KEPCO)', 'CO', 'CO2', 'r2(KEPCO)', 'etc']
    
    for i in range(4):
        df = df.replace(' '*(i+1), np.nan) # 공백제거
    
    return df
```

```python
df = load_data(inputPath)
```
