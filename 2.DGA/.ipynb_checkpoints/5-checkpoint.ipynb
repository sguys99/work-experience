```python
def return_r1(gas):
    if gas['long_term']==0: # 초기운전 변압기: 가압 후 3년 이내
        if (gas['H2']<=200) & (gas['C2H2']<=0.5) & (gas['C2H4']<=100) & \
           (gas['CH4']<=150) & (gas['C2H6']<=200) & (gas['C3H8']<=150) & \
           (gas['TCG2']<=500) & (gas['CO']<=400) & (gas['CO2']<=2500):
            return '정상'
        
        elif (400<gas['H2']) | (200<gas['C2H4']) | (250<gas['CH4'])|\
             (350<gas['C2H6']) | (250<gas['C3H8']) | (1000<gas['TCG2'])|\
             (800<gas['CO']) | (5000<gas['CO2']):
            return '이상'
        
        else:
            return '요주의'
        
    else:
        if (gas['H2']<=200) & (gas['C2H2']<=10) & (gas['C2H4']<=100) & \
             (gas['CH4']<=150)&(gas['C2H6']<=200) & (gas['C3H8']<=150) & \
             (gas['TCG2']<=500):
            
            if gas['dTCG2'] >= 200:
                return '요주의II'
            
            elif gas['dTCG2'] >= 100:
                return '요주의I'
            
            else:
                return '정상'
        
        elif (120 < gas['C2H2']) | (4000<gas['TCG2']):
            return '위험'
        
        elif (800<gas['H2']) | (60<gas['C2H2']<=120)|(500<gas['C2H4']) | \
             (750<gas['CH4'])|(750<gas['C2H6']) | (750<gas['C3H8']) |\
             (2500<gas['TCG2']<=4000):
            
            if gas['dTCG2'] >= 300:
                return '위험'
            else:
                return '이상'
            
        elif (400<gas['H2']<=800) | (20<gas['C2H2']<=60)|\
             (200<gas['C2H4']<=500) | (250<gas['CH4']<=750)|\
             (350<gas['C2H6']<=750) | (250<gas['C3H8']<=750) |\
             (1000<gas['TCG2']<=2500):
            
            if gas['dTCG2'] >= 200:
                return '이상'
            
            else:
                return '요주의II'
        
        else:
            return '요주의I'
```

```python
def return_r2(gas):
    if (gas['CO'] <= 800) & (gas['CO2'] <= 5000):
        return '정상'
    elif (gas['CO']>1200)|(gas['CO2'] > 7000):
        return '주의II'
    else :
        return '주의I'
```

```python
df['r1(func)'] = df[['long_term','H2', 'C2H2', 'C2H4', 'C2H6', 'C3H8',
                     'CH4','TCG2', 'CO', 'CO2', 'dTCG2']].apply(return_r1, axis = 1)
df['r2(func)'] = df[['CO', 'CO2']].apply(return_r2, axis = 1)
```

#### 5.2 IEEE 기준

1) Key gas method  
- **H2**, **CH4**(메탄), **C2H6**(에탄), **C2H4**(에틸렌), **C2H2**(아세틸렌), **CO** 6개 Key Gas의 비를 활용한 판정법
- **판정등급** : thermal(oil), thermal(cel), PD, Arc
    - **thermal(oil)** : Key gas - C2H4/ 절연유 분해로 C2H2, CH4 생성/ H, C2H6 미량 발생/ 상태가 심각해지 전기적 접점이 관련되면 C2H2이 생성되기 시작함
    - **thermal(cel)** : Key gas - CO / 절연지 과열로 CO, CO2가 다량 생성됨/ 과열로 인한 절연유가 스며들게 되면 CH4나 C2H4 확인되기도 함
    - **PD** : Key gas - H2 / 저에너지 방전으로 인해 H, CH4 생성/ 미량의 C2H6, C2H4 생성/ 절연지에 PD가 발생하면 CO, CO2가 발생하기도 함  
    - **Arc** : Key gas - C2H2 / H2, C2H2 다량 발생/ 미량의 CH4, C2H4/ 절연지에 발생하면 CO, CO2가 발생
- 참고자료 IEEE Std C57.104-2008, pp12~13

```python
H2_lim = 100.0; CH4_lim = 120.0; CO_lim = 350.0; C2H2_lim = 1.0;
C2H4_lim = 50.0; C2H6_lim = 65.0;
```

```python
TCG_lim = (H2_lim+CH4_lim+C2H2_lim+C2H4_lim)*2 + (CO_lim + C2H6_lim)
```

```python
#IEEE 기준에 자체 고안 개념 추가하여 작성
def return_keygas_method(gas, keygas_th = 0.85, TCG_th = 1370.):
    # TCG_th : iEEE Std pp15의 Doernenburg 방법 참고
    # keygas 비중이 IEEE의 th와 비교하여 85% 이상을 만족하면 인정
    # TCG1 = H2 + C2H2 + C2H4 + CH4 + C2H6
    # TCG_th = (H2_lim+CH4_lim+C2H2_lim+C2H4_lim)*2 + (CO_lim + C2H6_lim)
    concentration = gas['TCG1']*2 + gas['CO']-gas['C2H6']
    keygas_total = gas['TCG1'] + gas['CO']
    if concentration < TCG_th:
        return 'normal'
    
    else:
        if (gas['C2H4'] / keygas_total) > (0.63*keygas_th) :
            return 'thermal(oil)'
        
        elif (gas['CO']/keygas_total) > (0.92*keygas_th) :
            return 'thermal(cel)'
        
        elif (gas['H2']/keygas_total) > (0.85*keygas_th) :
            return 'PD'
        
        elif (gas['C2H2']/keygas_total) > (0.30*keygas_th) :
            return 'Arc'
        
        else :
            return 'NA' 
```

```python
df['r(keygas)'] = df[['C2H4', 'CO', 'H2', 'C2H6','C2H2', 
                      'TCG1']].apply(return_keygas_method, axis = 1)
```

```python
fig, ax = plt.subplots(figsize = (figure_hrz, 6))
df['r(keygas)'].value_counts(dropna = False).plot(kind = 'bar', rot = 45, fontsize = font_n)
for p in ax.patches:
    ax.text(p.get_x()+p.get_width()/2., p.get_height()+1000, str(int(p.get_height())), ha='center', fontsize = font_n)
ax.set_title('Number of DGA samples (by r(keygas) result)', fontsize = font_n)
# ax.set_xlabel('fetures', fontsize = 14)
ax.set_ylabel('counts', fontsize = font_n)
plt.show()
```

2) Doernenburg ratio method  
- **CH4/H2**, **C2H2/C2H4**(아세틸렌/에틸렌), **C2H2/CH4**(아세틸렌/메탄), **C2H6/C2H2**(에탄/아세틸렌), **C2H4/C2H6**(에틸렌/에탄) 5개 Gas Ratio를 활용한 판정법
    - **판정등급** : normal, thermal, PD, Arc, NA
- 참고자료 IEEE Std C57.104-2008, pp14~15

- 진단절차
    - 1) H2, CH4, C2H2, C2H4 중 하나라도 L1 * 2 보다 높은 값이 있는지 확인 > 없으면 정상
    - 2) 1)에 해당하면 CO, CH6 중 하나라도 L1을 초과하는지 확인 > 없으면 정상
    - 3) 2)에 해당하면 CH4, H2, C2H2, C2H4, C2H6 중 하나라도 L1보다 큰지 확인 > 없으면 N.A.  
    - 4) R1, R2, R3, R4를 계산하여 상태를 판정

```python
def return_doernburg(gas):
    L1 = {'H2': 100, 'CH4': 120, 'CO': 350, 'C2H2': 1, 'C2H4': 50, 'C2H6': 65}
    
    test = 'Invalid'
    
    if (gas['H2'] <= 2*L1['H2'])|(gas['CH4'] <= 2*L1['CH4'])| \
       (gas['CO'] <= 2*L1['CO'])|(gas['C2H2'] <= 2*L1['C2H2'])| \
       (gas['C2H4'] <= 2*L1['C2H4'])|(gas['C2H6'] <= 2*L1['C2H6']):
        return 'normal'
    
    elif (gas['CO'] <= L1['CO'])&(gas['C2H6'] <= L1['C2H6']):
        return 'normal'
    
    elif ((gas['CH4'] <= L1['CH4'])|(gas['H2'] <= L1['H2']))& \
          ((gas['C2H2'] <= L1['C2H2'])|(gas['C2H4'] <= L1['C2H4']))&\
          ((gas['C2H2'] <= L1['C2H2'])|(gas['CH4'] <= L1['CH4']))&\
         ((gas['C2H6'] <= L1['C2H6'])|(gas['C2H2'] <= L1['C2H2']))&\
         ((gas['C2H4'] <= L1['C2H4'])|(gas['C2H6'] <= L1['C2H6'])):
        return 'NA'
    
    else :
        test = 'valid'
    
    
    R1 = gas['CH4/H2']; R2 =gas['C2H2/C2H4']; R3 =gas['C2H2/CH4']; 
    R4 = gas['C2H6/C2H2']; R5 = gas['C2H4/C2H6']; # limit concetration 반영할지 검토요(4.22)   

    if test == 'valid':
        if R1 <0.1:
            if (R3 < 0.3)&(R4 > 0.4):
                return 'PD'
            else :
                return 'NA'
            
        elif ( 0.1 <= R1 <= 1): # 등호 수정완료(4.22)
            if (R2>=0.75)&(R3>=0.3)&(R4<=0.4):
                return 'Arc'
                    
            else:
                return 'NA'
                
        else: # if (R1 > 1)
            if (R2<0.75)&(R3<0.3)&(R4>0.4):
                return 'thermal'
            
            else:
                return 'NA'
        
    else:
        return 'NA'
```

```python
df['r(doernburg)'] = df.apply(return_doernburg, axis = 1)
```

3) Rogers ratio method  
- **CH4/H2**, **C2H2/C2H4**(아세틸렌/에틸렌), **C2H2/CH4**(아세틸렌/메탄), **C2H6/C2H2**(에탄/아세틸렌), **C2H4/C2H6**(에틸렌/에탄) 5개 Gas Ratio를 활용한 판정법
    - **판정등급** : normal, thermal1(low temp), thermal2(<700'C), thermal3(>700'C), PD, Arc
- 참고자료 IEEE Std C57.104-2008, pp16~17

```python
def return_rogers(gas):
    R1 = gas['CH4/H2']; R2 =gas['C2H2/C2H4']; R5 = gas['C2H4/C2H6'];
    
    if R2 < 0.1:
        if 0.1 < R1 < 1.0:
            if R5 < 1.0:
                return 'normal'
            
            else: # 1.0 <R5 <3.0
                return 'thermal1'
            
        elif R1 >= 1.0:
            if 1.0 < R5 < 3.0:
                return 'thermal2'
            
            else: # R5 > 3.0
                return 'thermal3'
            
        else: # R1 < 0.1
            if R5 < 1.0:
                return 'PD'
            else:
                return 'NA'
        
    else: # 0.1<R2<3.0
        if 0.1<R1<1.0:
            if R5 > 3.0:
                return 'Arc'
            else:
                return 'NA'
            
        else:
            return 'NA'
```

```python
df['r(rogers)'] = df.apply(return_rogers, axis = 1)
```

```python
PERCENTILE = 90

C2H2_nor_lim = np.percentile(df['C2H2'].values, PERCENTILE, 
                                 interpolation = 'nearest')
C2H4_nor_lim = np.percentile(df['C2H4'].values,  PERCENTILE,
                                 interpolation = 'nearest')
CH4_nor_lim = np.percentile(df['CH4'].values,  PERCENTILE,
                                 interpolation = 'nearest')
H2_nor_lim = np.percentile(df['H2'].values,  PERCENTILE,
                                 interpolation = 'nearest')
C2H6_nor_lim = np.percentile(df['C2H6'].values,  PERCENTILE,
                                 interpolation = 'nearest')
print('Typical values:')
print('C2H2: {}'.format(C2H2_nor_lim))
print('C2H4: {}'.format(C2H4_nor_lim))
print('CH4: {}'.format(CH4_nor_lim))
print('H2: {}'.format(H2_nor_lim))
print('C2H6: {}'.format(C2H6_nor_lim))
```

```python
def val_gas(gas): #gas 측정 값이 가 계산이 가능할 정도인지 계산
    H2_lim = 5.0; CH_lim = 1.0;
    if gas['C2H2'] <=CH_lim*10:
        C2H2_val = CH_lim
    else:
        C2H2_val = gas['C2H2']
    
    if gas['C2H4'] <=CH_lim*10:
        C2H4_val = CH_lim
    else:
        C2H4_val = gas['C2H4']
        
    if gas['CH4'] <=CH_lim*10:
        CH4_val = CH_lim
    else:
        CH4_val = gas['CH4']
        
    if gas['C2H6'] <=CH_lim*10:
        C2H6_val = CH_lim
    else:
        C2H6_val = gas['C2H6']
        
    if gas['H2'] <= H2_lim*10:
        H2_val = H2_lim
    else:
        H2_val = gas['H2']
        
    return C2H2_val, C2H4_val, CH4_val, H2_val, C2H6_val
```

```python
def return_IEC1(gas, C2H2_nor_lim, C2H4_nor_lim, CH4_nor_lim, H2_nor_lim, C2H6_nor_lim):

    # key gas :'C2H2', 'C2H4', 'CH4', 'H2', 'C2H6'
    if (gas['C2H2']<= C2H2_nor_lim) and (gas['C2H4']<= C2H4_nor_lim) and \
       (gas['CH4']<= CH4_nor_lim) and (gas['H2']<= H2_nor_lim) and \
       (gas['C2H6']<= C2H6_nor_lim): #C2H2_nor_lim=0 이므로 등호포함
        return 'normal'
    
    else:
        C2H2, C2H4, CH4, H2, C2H6 = val_gas(gas)
        r1 = C2H2/C2H4; r2 = CH4/H2; r3 = C2H4/C2H6;
        
        if (r2 < 0.1) & (r3 < 0.2):
            return 'PD'
        
        elif (r1 > 1.0) & (0.1<= r2 <= 0.5) & (r3 > 1.0):
            return 'D1'
        
        elif (0.6 <= r1 <= 2.5) & (0.1<= r2 <= 1.0) & (r3 > 2.0):
            return 'D2'
        
        elif (r2 > 1.0) & (r3 < 1.0):
            return 'T1'
        
        elif (r1 < 0.1) & (r2 > 1.0) & (1.0 <= r3 <= 4.0):
            return 'T2'
        
        elif (r1 < 0.2) & (r2 > 1.0) & (r3 > 4.0):
            return 'T3'
        
        else:
            return 'NA'
```

```python
df['r(IEC1)'] = df.apply(return_IEC1, 
                        C2H2_nor_lim = C2H2_nor_lim, C2H4_nor_lim= C2H4_nor_lim, 
                        CH4_nor_lim= CH4_nor_lim, H2_nor_lim=H2_nor_lim, 
                        C2H6_nor_lim=C2H6_nor_lim, axis = 1)
```

2) Duval method

```python
def return_IEC2(gas, C2H2_nor_lim, C2H4_nor_lim, CH4_nor_lim, H2_nor_lim, C2H6_nor_lim):

    if (gas['CH4']>120) or (gas['C2H4']>50) or (gas['C2H2']>1):
        return 'None'
    # key gas :'C2H2', 'C2H4', 'CH4', 'H2', 'C2H6'
    elif (gas['C2H2']<= C2H2_nor_lim) and (gas['C2H4']<= C2H4_nor_lim) and \
       (gas['CH4']<= CH4_nor_lim) and (gas['H2']<= H2_nor_lim) and \
       (gas['C2H6']<= C2H6_nor_lim): #C2H2_nor_lim=0 이므로 등호포함
        return 'Normal'
    
    else:
        C2H2, C2H4, CH4, _, _ = val_gas(gas)
        total = (C2H2 + C2H4 + CH4)
        p1 = (C2H2/total)*100.0; p2 = (C2H4/total)*100.0; p3 = (CH4/total)*100.0;
        
        if p1 >= 98:
            return 'PD'
        
        elif p3 < 4:
            if 20 <= p2 <= 50:
                return 'T2'
            elif p2<20:
                return 'T1'
            else:
                return 'T'
        
        elif (p3 < 13) & (p2 < 23):
            return 'D1'
        
        elif p3 < 15 :
            if p2 >= 50:
                return 'T3'
            else:
                return 'T'
        
        elif (29 <= p3 <= 77) & (23 <= p2 <= 40):
            return 'D2'
        
        else:
            return 'T'
```

```python
df['r(IEC2)'] = df[['C2H2', 'C2H4', 'CH4', 'H2', 'C2H6']].apply(return_IEC2, 
                        C2H2_nor_lim = C2H2_nor_lim, C2H4_nor_lim= C2H4_nor_lim, 
                        CH4_nor_lim= CH4_nor_lim, H2_nor_lim=H2_nor_lim, 
                        C2H6_nor_lim=C2H6_nor_lim, axis = 1)
```

```python
fig, ax = plt.subplots(figsize = (figure_hrz, 6))
df['r(IEC2)'].value_counts(dropna = False).plot(kind = 'bar', rot = 45, fontsize = font_n)
for p in ax.patches:
    ax.text(p.get_x()+p.get_width()/2., p.get_height()+1000, str(int(p.get_height())), ha='center', fontsize = font_n)
ax.set_title('Number of DGA samples (by r(IEC2) result)', fontsize = font_n)
# ax.set_xlabel('fetures', fontsize = 14)
ax.set_ylabel('counts', fontsize = font_n)
plt.show()
```
